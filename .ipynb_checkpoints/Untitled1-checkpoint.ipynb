```python
import pandas as pd
import matplotlib.pyplot as plt

# read data from CSV file
df = pd.read_csv("falling_15.csv")

# calculate magnitude of acceleration and gyroscope readings
df['accel_mag'] = (df['accel_x']**2 + df['accel_y']**2 + df['accel_z']**2)**0.5
df['gyro_mag'] = (df['gyro_x']**2 + df['gyro_y']**2 + df['gyro_z']**2)**0.5

# plot acceleration and gyroscope magnitude vs. sample number
plt.plot(df.index, df['accel_mag'], label='Acceleration Magnitude')
plt.plot(df.index, df['gyro_mag'], label='Gyroscope Magnitude')
plt.xlabel('Sample Number')
plt.ylabel('Magnitude')
plt.legend()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# read data from CSV files
df1 = pd.read_csv('falling_1.csv')
df2 = pd.read_csv('falling_2.csv')

# calculate magnitudes of acceleration and gyro readings
df1['accel_mag'] = (df1['accel_x']**2 + df1['accel_y']**2 + df1['accel_z']**2)**0.5
df1['gyro_mag'] = (df1['gyro_x']**2 + df1['gyro_y']**2 + df1['gyro_z']**2)**0.5

df2['accel_mag'] = (df2['accel_x']**2 + df2['accel_y']**2 + df2['accel_z']**2)**0.5
df2['gyro_mag'] = (df2['gyro_x']**2 + df2['gyro_y']**2 + df2['gyro_z']**2)**0.5

# plot acceleration magnitudes
plt.plot(df1.index, df1['accel_mag'], label='File 1')
plt.plot(df2.index, df2['accel_mag'], label='File 2')

plt.title('Acceleration Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.legend()
plt.show()

# plot gyro magnitudes
plt.plot(df1.index, df1['gyro_mag'], label='File 1')
plt.plot(df2.index, df2['gyro_mag'], label='File 2')

plt.title('Gyro Magnitude')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# list of file names
file_names = ['falling_1.csv', 'falling_2.csv', 'falling_3.csv', 'falling_4.csv', 'falling_5.csv', 'falling_6.csv', 'falling_7.csv', 'falling_8.csv', 'falling_9.csv', 'falling_10.csv']

# read data from each file and create a dataframe
dfs = []
for file in file_names:
    df = pd.read_csv(file)
    df['magnitude'] = (df['accel_x']**2 + df['accel_y']**2 + df['accel_z']**2 + df['gyro_x']**2 + df['gyro_y']**2 + df['gyro_z']**2)**0.5
    dfs.append(df)



    
# concatenate all dataframes into one
df_concat = pd.concat(dfs)

# plot magnitude vs index
plt.plot(df_concat.index, df_concat['magnitude'])
plt.title('Magnitude vs Index')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# define file names
file_names = ['falling_1.csv', 'falling_2.csv', 'falling_3.csv', 'falling_4.csv', 'falling_5.csv', 'falling_6.csv', 'falling_7.csv', 'falling_8.csv', 'falling_9.csv', 'falling_10.csv']

# create an empty list to store dataframes
dfs = []

# loop through the files and read them into dataframes
for file in file_names:
    df = pd.read_csv(file)
    # calculate magnitude for accelerometer data
    df['accel_mag'] = (df['accel_x']**2 + df['accel_y']**2 + df['accel_z']**2)**0.5
    # calculate magnitude for gyro data
    df['gyro_mag'] = (df['gyro_x']**2 + df['gyro_y']**2 + df['gyro_z']**2)**0.5
    dfs.append(df)

# create subplots with 2 rows and 1 column
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# loop through the dataframes and plot the magnitudes on the corresponding subplot
for i, df in enumerate(dfs):
    axs[0].plot(df.index, df['accel_mag'], label=f'File {i+1}')
    axs[1].plot(df.index, df['gyro_mag'], label=f'File {i+1}')

# set x and y axis labels
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Accelerometer Magnitude')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Gyro Magnitude')

# set the title for the entire figure
fig.suptitle('Accelerometer and Gyro Magnitudes for Multiple Files')

# add legend to the subplots
axs[0].legend()
axs[1].legend()

# show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# define file names
file_names = ['walking_1.csv', 'walking_2.csv', 'walking_3.csv', 'walking_4.csv', 'walking_5.csv', 'walking_6.csv', 'walking_7.csv', 'walking_8.csv', 'falling_9.csv', 'falling_10.csv']

# create an empty list to store dataframes
dfs = []

# loop through the files and read them into dataframes
for file in file_names:
    df = pd.read_csv(file)
    # calculate magnitude for accelerometer data
    df['accel_mag'] = (df['accel_x']**2 + df['accel_y']**2 + df['accel_z']**2)**0.5
    # calculate magnitude for gyro data
    df['gyro_mag'] = (df['gyro_x']**2 + df['gyro_y']**2 + df['gyro_z']**2)**0.5
    dfs.append(df)

# create subplots with 2 rows and 1 column
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# loop through the dataframes and plot the magnitudes on the corresponding subplot
for i, df in enumerate(dfs):
    axs[0].plot(df.index, df['accel_mag'], label=f'File {i+1}')
    axs[1].plot(df.index, df['gyro_mag'], label=f'File {i+1}')

# set x and y axis labels
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Accelerometer Magnitude')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Gyro Magnitude')

# set the title for the entire figure
fig.suptitle('Accelerometer and Gyro Magnitudes for Multiple Files')

# add legend to the subplots
axs[0].legend()
axs[1].legend()

# show the plot
plt.show()
```
